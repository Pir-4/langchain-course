```python
from langsmith import traceable
import random
import time


@traceable
def generate_random_number():
    return random.randint(0, 100)

@traceable
def generate_string_delay(input_str:str):
    number = random.randint(1, 5)
    time.sleep(number)
    return  f"{input_str} ({number})"

@traceable
def random_error():
    number = random.randint(0, 1)
    if number == 0:
        raise  ValueError("Random error")
    else:
        return "No error"
```

```python
for _ in range(10):
    generate_random_number()
    generate_string_delay("Smth StRing")
    try:
        random_error()
    except ValueError:
        pass
```

```python
@traceable(name="Chitcat Marker")
def error_generation_func(question: str):
    dalay = random.randint(0, 3)
    time.sleep(dalay)
    number = random.randint(0, 1)
    if number == 0:
        raise  ValueError("Random error")
    else:
        return llm.invoke(question)
```

```python
for _ in range(3):
    try:
        error_generation_func("How are you today?")
    except ValueError:
        pass
```
